```python
import numpy as np
import matplotlib.pyplot as plt
```

# This notebook goes through the basics of Linear Algebra

```python
""" Norm is denoted as |v| or ||v|| (l2 norm) and represents the vector's magnitude (length) """
vector_v = np.array([3, 4])
vector_v_sum = 0
for i in vector_v:
    vector_v_sum += i**2
vector_v_length = np.sqrt(vector_v_sum)
print(vector_v_length)
```

```python
""" Eucleidian Norm is to calculate the distance between Vector A and Vector B.
d(A, B) = SQRT(SUM((Ai - Bi)^2))
"""
A = [1, 2]
B = [4, 6]

vector_sum = 0
for a, b in zip(A, B):
    vector_sum += (b - a)**2

eucleidian_length = np.sqrt(vector_sum)

plt.figure(1)
plt.arrow(0, 0, A[0], A[1], color='blue')
plt.arrow(0, 0, B[0], B[1], color='green')
plt.arrow(A[0], A[1], B[0] - A[0], B[1] - A[1], color='red')
print(eucleidian_length)
```

```python
A = set([1, 3, 5])
B = set([4, 5, 6])
SIZE = len(A) + len(B)

p_over = len(A & B) / SIZE
p_under = len(B) / SIZE

PA = p_over / p_under
print(PA)
```
