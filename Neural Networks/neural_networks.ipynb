# Neural Networks
**This is an introduction to what neural networks are and how they work.**

Neural networks can be looked at as graphs, where we map a set of neurons in **input layer** to an **output layer**. In some cases, the **input layer** from now on called $U_{in}$ can also be the **output layer** called $U_{out}$. A network can also have multiple layers between the $U_{in}$ and $U_{out}$, which are called for **hidden layers** or $U_{hidden}$. Each layer contains one or more neurons. There are some constraints that we must follow: <br>

$$U_{in} \neq \phi$$
$$U_{out} \neq \phi$$
$$U_{hidden} \cap (U_{in} \cup U_{out}) = \phi$$

Each neuron carries a set of **Weights** which has the length of the predecessor set. And three quantities:<br>
**Network Input** $net_u$<br>
**Activation** $act_u$<br>
**Output** $out_u$<br>

$$net_u = b + \sum_{i=1}w_ix_i$$
$$act_u = \sigma(net_u)$$
$$out_u = act_u$$

The $\sigma$ is an activation function which will apply a non-linearity to the neuron. Meaning the network is capable of doing classifications etc, where we cannot draw a straight line through each bucket. There are multiple activation functions to use. However, most common is **ReLU**, **Sigmoid**, **Tanh** to name a few.

$$Sigmoid = \sigma(x) = \dfrac{1}{1 + e^{-x}}$$
$$ReLU = \sigma(x) = \begin{cases} 0: x <= 0 \\ x: x > 0 \end{cases}$$
$$Tanh = \sigma(x) = \dfrac{e^x - e^{-x}}{e^x + e^{-x}}$$



```python
from random import uniform

""" Imports """
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
""" Activation Function Demo """
X = np.arange(-5, 5, 0.1)

def sigmoid(x) -> float:
    return 1 / (1 + np.exp(-x))

def relu(x) -> float:
    return x * (x > 0)

def tanh(x) -> float:
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
fig.tight_layout(h_pad=2.0)

Y = sigmoid(X)
ax1.plot(X, Y)
ax1.set_title("Sigmoid")

Y = relu(X)
ax2.plot(X, Y)
ax2.set_title("ReLU")

Y = tanh(X)
ax3.plot(X, Y)
ax3.set_title("Tanh")
```

```python
class Neuron:
    def __init__(self, input_size: int) -> None:
        self.bias = uniform(-1, 1)
        self.weights = np.arange(0, 0, input_size)
    
    def forward(self, input: list[float]) -> float:
        net = self.weights.dot(input) + self.bias
        act = tanh(net)
        out = act
        return out
```
